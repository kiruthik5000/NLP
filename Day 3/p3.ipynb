```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
df = pd.read_csv("./data/p1.csv", parse_dates=['Date'])
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)

y = df['Close_diff'].dropna()
y.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
plot_acf(y, lags=30, ax=ax[0])
ax[0].set_ylabel("ACF Values")
ax[0].set_xlabel("Lags")
ax[0].set_title("ACF")
plot_pacf(y, lags=40, ax=ax[1], color='red')
ax[1].set_xlabel("PACF Values")
ax[1].set_ylabel("Lags")
ax[1].set_title("PACF")
plt.show()
```
