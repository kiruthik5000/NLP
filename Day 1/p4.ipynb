```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import  seasonal_decompose
```

```python
df = pd.read_csv("./Data/p4.csv", parse_dates=['Date'])
df = df.sort_values('Date')
df.set_index('Date', inplace=True)
df.head()
close = df['Close']
close = close.interpolate()
```

```python
def plot_decompose(decmp, name):
    plt.figure(figsize=(20, 16))

    plt.subplot(411)
    plt.plot(decmp.observed)
    plt.title(name + " Decompose: Observed")
    plt.subplot(412)
    plt.plot(decmp.trend)
    plt.title(name + " Decompose: Trends")
    plt.subplot(413)
    plt.plot(decmp.seasonal)
    plt.title(name + " Decompose: Seasonal")
    plt.subplot(414)
    plt.plot(decmp.resid)
    plt.title(name + " Decompose: Residual")
    plt.show()
```

```python
add_decompose = seasonal_decompose(
    close,
    model='additive',
    period=365
)

mul_decompose = seasonal_decompose(
    close,
    model='multiplicative',
    period=365
)

plot_decompose(decmp=add_decompose, name="Additive")
plot_decompose(decmp=mul_decompose, name="Multiplicative")
```
