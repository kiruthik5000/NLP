```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/c1.csv")
```

```python
df = df.dropna()
df['Datetime'].astype('datetime64[ns]')
series = df['Power_Consumption_diff']
```

```python
diff_series = series.diff().dropna()
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plt.figure(figsize=(6, 4))
plot_acf(diff_series, lags=30)
plt.title("ACF Plot of Differenced Power Consumption")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plot_pacf(diff_series, lags=30)
plt.title("PACF")
plt.xlabel("Lags")
plt.ylabel("PACF Values")
plt.show()
```
