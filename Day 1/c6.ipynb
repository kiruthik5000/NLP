```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Data/c4.csv", parse_dates=['DATE'])
df.set_index("DATE", inplace=True)
df.head()
```

```python
df_diff = df['Consumption'].diff().dropna()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_diff)
plt.title("Difference Power Consumption")
plt.xlabel("Year")
plt.ylabel("Difference Power Consumption")
plt.show()
```
