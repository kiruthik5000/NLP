```python
from hashlib import file_digest

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, acf
```

```python
df = pd.read_csv("./Data/c4.csv", parse_dates=['DATE'])
df.set_index('DATE', inplace=True)

series = df['Consumption'].interpolate()
```

```python
lags = 400
acf_values = acf(series, nlags=lags)
```

```python
plt.figure(figsize=(10, 6))

# Continuous ACF line
plt.plot(range(len(acf_values)), acf_values, linewidth=1)

# Zero line
plt.axhline(0, color='black', linewidth=1)

# Confidence bounds (95%)
conf = 1.96 / np.sqrt(len(series))
plt.axhline(conf, color='gray', linestyle='--')
plt.axhline(-conf, color='gray', linestyle='--')

plt.title("Autocorrelation Plot", fontsize=14)
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.grid(True)

plt.show()
```
