```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, acf
```

```python
df = pd.read_csv("data/p1.csv", parse_dates=['Date'])
df.set_index('Date')
y = df['Close']
y.dropna(inplace=True)
y.head()
```

```python
# differenced data
diff = y.diff().dropna()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
plot_pacf(y, lags=30, ax=ax[0], color='blue')
ax[0].set_title("ACF of Differenced data")
ax[0].set_xlabel("Lags")
ax[0].set_ylabel("ACF Values")
plot_pacf(y, lags=30, ax=ax[1], color='red')
ax[1].set_title("PACF of Differenced data")
ax[1].set_xlabel("Lags")
ax[1].set_ylabel("PACF Values")
plt.show()
```
